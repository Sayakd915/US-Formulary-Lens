```python
## Changing January text data to csv format
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Dataset-CodeFest '25/Jan_formulary_data.txt", sep="|")
df.to_csv("/content/drive/MyDrive/Dataset-CodeFest '25/Jan_formulary_data.csv", index = False)
df.head(20)
```

```python
df["TIER_LEVEL_VALUE"] = df["TIER_LEVEL_VALUE"].astype(int)
df["QUANTITY_LIMIT_AMOUNT"] = df["QUANTITY_LIMIT_AMOUNT"]
df["QUANTITY_LIMIT_DAYS"] = df["QUANTITY_LIMIT_DAYS"]
```

```python
df_cleaned = df.dropna(subset=["TIER_LEVEL_VALUE", "QUANTITY_LIMIT_AMOUNT", "QUANTITY_LIMIT_DAYS"])
df_sampled = df_cleaned.sample(n=5000, random_state=42)
df_sampled.head(20)
```

```python
import numpy as np
df_sampled.replace(r'^\s*$', np.nan, regex=True, inplace=True)
df_sampled["QUANTITY_LIMIT_AMOUNT"] = pd.to_numeric(df["QUANTITY_LIMIT_AMOUNT"], errors="coerce").fillna(0)
df_sampled["QUANTITY_LIMIT_DAYS"] = pd.to_numeric(df["QUANTITY_LIMIT_DAYS"], errors="coerce").fillna(0)
df_sampled.head(20)
```

```python
##Distribution along Tier Levels

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

plt.figure(figsize=(8, 5))
sns.countplot(x="TIER_LEVEL_VALUE", data=df_sampled, palette="viridis")
plt.title("Distribution of Drugs by Tier Level")
plt.xlabel("Tier Level")
plt.ylabel("Count")
plt.show()
```

```python
##Quantity Limit Amounts vs Days
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df_sampled["QUANTITY_LIMIT_AMOUNT"], y=df_sampled["QUANTITY_LIMIT_DAYS"], alpha=0.5)
ax = plt.gca()
ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=10))
plt.xticks(rotation=45)
plt.xlabel("Quantity Limit Amount")
plt.ylabel("Quantity Limit Days")
plt.title("Quantity Limit Days vs. Amount")

plt.show()
```
